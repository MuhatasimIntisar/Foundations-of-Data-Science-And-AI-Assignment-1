```python
import pandas as pd
from scipy import stats

# Load dataset
df = pd.read_csv("C:\\Users\\sirit\\Downloads\\GymMembers.csv")

# Group calories burned by workout type
cardio = df[df["Workout_Type"] == "Cardio"]["Calories_Burned"]
hiit = df[df["Workout_Type"] == "HIIT"]["Calories_Burned"]
strength = df[df["Workout_Type"] == "Strength"]["Calories_Burned"]
yoga = df[df["Workout_Type"] == "Yoga"]["Calories_Burned"]

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(cardio, hiit, strength, yoga)

# Display results
print("F-statistic:", f_statistic)
print("p-value:", p_value)

# Group means
print("\nAverage calories burned per workout type:")
print(df.groupby("Workout_Type")["Calories_Burned"].mean())
```
